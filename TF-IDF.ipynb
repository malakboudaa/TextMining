```python
import pandas as pd
import re
#importation de la librairie qui permet d'utiliser tfidf
from sklearn.feature_extraction.text import TfidfVectorizer  

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

StopWords = stopwords.words("english")
StopWords

DocA = "Data with many fields offer greater statistical power, while data with higher complexity may lead to a higher false discovery rate. Big data analysis challenges include capturing data, data storage, data analysis, search, sharing, transfer, visualization, querying, updating, information privacy, and data source"
DocB = "The analysis of big data presents challenges in sampling, and thus previously allowing for only observations and sampling. Therefore, big data often includes data with sizes that exceed the capacity of traditional software to process within an acceptable time and value"
```

```python
#creer une instance de la classe tfidfvectorizer
#^pur convertir les documents (docA / docB) en un matrice tf_idf
vectorizer = TfidfVectorizer()
#transformer les doc en vecteur
vectors = vectorizer.fit_transform([DocA, DocB])
#recuperer les nom correspondant des caracteristiques
feature_names = vectorizer.get_feature_names_out()
```

```python
#convertise la matrice vectors en une matricedense
dense = vectors.todense()
#convertise la matrice dense en liste, 
#Chaque liste interne représente une ligne de la matrice
denseList = dense.tolist()
#creer un data frame 
df = pd.DataFrame(denseList, columns=feature_names)  # Corrected variable name
df
```
