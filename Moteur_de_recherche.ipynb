```python
import pandas as pd
import re
#importation de la librairie qui permet d'utiliser tfidf
from sklearn.feature_extraction.text import TfidfVectorizer  

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

StopWords = stopwords.words("english")
```

```python
#enelver les stopwords
def clean_sentence(sen):
    sen = re.sub("\d+|\.|,|Md$|%|:|$|\)|\(|\n", '', sen).lower()
    
    senList = word_tokenize(sen.lower())
    DocFil = [w for w in senList if not w in stopwords.words('english')]
    senFil = ' '.join(DocFil)
    return senFil
```

```python
vectorizer = TfidfVectorizer()
```

```python
data.abstract.values #recuperer que les abstract
```

```python
#transformer les abstract en liste (chaine de caractere néttoyés)
#tout en minuscule et enelver les stopwords
abstractList = [abs for abs in list(data.abstract.values)] 
#abstractList
```

```python
vectors = vectorizer.fit_transform(abstractList)
feature_names = vectorizer.get_feature_names_out()
```

```python
dense = vectors.todense()
denseList = dense.tolist()
df = pd.DataFrame(denseList, columns = feature_names)
df
```

```python
query = "education best system africa"
```

```python
#changer le nom des colonnes
engine = df.transpose()
colnames = ["chapitre_"+str(i+1) for i in range(20)]
engine.columns = colnames
engine.head()
```

```python
word_tokenize(clean_sentence(query))
```

```python
#les tf/idf de chaque mot dans les document 
queryData =  engine.loc[word_tokenize(clean_sentence(query))]
tfidf = queryData.sum()
tfidf.sort_values(ascending= False).head(10)
```

```python
def book_engine(query, engine): 
    query_tokens = word_tokenize(clean_sentence(query.lower()))

    # Filtrer les tokens qui ne sont pas dans le DataFrame 
    valid_tokens = [token for token in query_tokens if token in engine.index]

    # retourner un resultat vide dans le cas ou les tokens sont pas valide
    if not valid_tokens:
        print("No valid tokens in the query.")
        return pd.Series()

    queryData = engine.loc[valid_tokens]
    tfidf = queryData.sum()
    tfidf_sorted = tfidf.sort_values(ascending=False).head(10)
    return tfidf_sorted
```

```python
query = "music enconomy east"
engine = engine
book_engine(query,engine)
```
